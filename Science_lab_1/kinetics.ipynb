```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['text.usetex'] = True
```

```python
dval = np.array([2.9,2.1,1.6,1.1,0.9,0.7,0.5])# enter Vn values for 1%
dval
```

```python
xval = [180*i for i in range(1,8)]
xval
```

```python
plotx = xval
ploty = np.log(dval[0]/dval)
print(plotx,ploty)
```

```python
(m,c) = np.linalg.lstsq(np.vstack([plotx, np.ones(len(plotx))]).T, ploty, rcond=None)[0]
print(m,c)
```

```python
plt.figure(figsize = (8,11))
plt.plot(np.linspace(min(plotx),max(plotx),10),m*np.linspace(min(plotx),max(plotx),10)+c)
plt.scatter(plotx,ploty,c = 'r')
# plt.title(r"\log\left(\frac{V_n}{V_0}\right) \text{ vs. time for 1\% FeCl}_3")
plt.title(r'$\log\left(\frac{V_n}{V_0}\right)$ vs. Time  for 1\% FeCl$_3$')

plt.xlabel(r'$\Delta t_n$ (in seconds)')
plt.ylabel(r"$log\left(\frac{V_n}{V_0}\right)$")
for xy in zip(plotx, ploty):
   plt.annotate('(%.0f, %.5f)' % xy, xy=xy)
plt.grid(True,color='g')
plt.savefig("kin1.png")
```

```python
p = m*2.303
p
```
