```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
array = [3.4,3.7,10.4,17.1,47.2,100]
print(2-np.log10(28.5))
```

```python
xval = [20*i for i in range(0,6)]
xval
```

```python
yval = 2-np.log10(array)[::-1]
yval
```

```python
plotx = xval
ploty = yval
print(plotx,ploty)
```

```python
(m,c) = np.linalg.lstsq(np.vstack([plotx, np.ones(len(plotx))]).T, ploty, rcond=None)[0]
print(m,c)
```

```python
plt.figure(figsize = (8,11))
plt.plot(np.linspace(min(plotx),max(plotx),10),m*np.linspace(min(plotx),max(plotx),10)+c)
plt.scatter(plotx,ploty,c = 'r')
plt.title(label = "Calibration graph")
plt.xlabel("Concentration(as a percentage of N/5000)")
plt.ylabel("Absorbance")
plt.scatter([(2-np.log10(28.5)-c)/m],[2-np.log10(28.5)],c = 'black')
for xy in zip(plotx, ploty):
   plt.annotate('(%.1f, %.3f)' % xy, xy=xy)

plt.annotate(text = '(31.706,0.545)',xy = (31.706,0.545))
plt.grid(True,color='g')
plt.savefig("trial4.png")
```
