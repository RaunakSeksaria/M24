```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
bval = np.array([2.66,2.42,1.88,1.33])*(10**(-5))
bval
```

```python
r = 0.118
xar = np.array([0,0.03,0.06,0.09])
xval = xar*xar + (r**2)
xval
```

```python
plotx = np.log(xval)
ploty = np.log(bval)
print(plotx,ploty)
```

```python
(m,c) = np.linalg.lstsq(np.vstack([plotx, np.ones(len(plotx))]).T, ploty, rcond=None)[0]
print(m,c)
```
