```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
dval = np.array([172,165,154,139,122,98,74,60.5,50.4,51.6,57.5,73.2,93.4,111,127,142,153,161,162,158,151,139,124,104,80.8,63.4,53.2,53,64,82.3,102,121,140,156,167,171])
dval
```

```python
xval = np.array([10*i for i in range(0,36)],dtype=float)
xval
```

```python
xval2 = (np.cos((np.pi/180)*xval))**2
xval2
```

```python
plotx = xval2[:9]
ploty = dval[:9]
print(len(ploty))
print(plotx,ploty)
```

```python
(m,c) = np.linalg.lstsq(np.vstack([plotx, np.ones(len(plotx))]).T, ploty, rcond=None)[0]
print(m,c)
```

```python
plt.figure(figsize = (8,11))
plt.plot(np.linspace(min(plotx),max(plotx),10),m*np.linspace(min(plotx),max(plotx),10)+c)
plt.scatter(plotx,ploty,c = 'r')
plt.title(label = "Malu's Law verification")
plt.xlabel("cos^2 theta")
plt.ylabel("Voltage(in mV)")
for xy in zip(plotx, ploty):
   plt.annotate('(%.2f, %.1f)' % xy, xy=xy)
plt.grid(True,color='g')
plt.savefig("malus_straight.png")
```
