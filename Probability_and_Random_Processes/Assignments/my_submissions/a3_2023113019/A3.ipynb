```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
number_of_iterations=10000
pmf=np.zeros(101)
pry=np.zeros(101)
for i in range(0,number_of_iterations):
    heads=0
    tails=0
    X=0
    Y=0
    for j in range(1,101):
        outcome=np.random.randint(0,2)
        if outcome == 0:
            tails+=1
        if outcome==1:
            heads+=1
        if (heads>tails):
            X+=1
        if heads==tails:
            Y+=1
    pmf[X]+=1
    pry[Y]+=1
pmf/=number_of_iterations
xval=np.array([i for i in range(101)])
```

```python
plt.figure(figsize = (20,11))
plt.bar(xval,pmf,color = 'g') 
plt.title("PMF-X")
plt.xlabel("Value of X")
plt.ylabel("P of X")
```

```python
ey=0
ez=0
pry = pry/number_of_iterations
for i in range(101):
    ey+=i*pry[i]
    ez+=i*pmf[i]

plt.figure(figsize = (20,11))
plt.bar(xval,pry,color = 'g')
plt.title("PMF - Y")
plt.ylabel("P of Y")
plt.xlabel("Value of Y")
print("Expectation of Y :",ey)
print("Expectation of Z :",ez)
```
