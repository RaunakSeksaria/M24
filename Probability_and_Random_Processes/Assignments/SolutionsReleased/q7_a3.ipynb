```python
import numpy as np
import matplotlib.pyplot as plt

num_simulations = 10000 
num_tosses = 100 

X_values = []

for _ in range(num_simulations):
    heads = 0
    tails = 0
    more_heads_count = 0
    
    for _ in range(num_tosses):
        toss = np.random.choice(['H', 'T'])  
        
        if toss == 'H':
            heads += 1
        else:
            tails += 1
        
        if heads > tails:
            more_heads_count += 1
    
    X_values.append(more_heads_count)

X_values = np.array(X_values)

pmf_X = np.bincount(X_values) / num_simulations

plt.figure(figsize=(10, 6))
plt.bar(range(len(pmf_X)), pmf_X, color='blue', alpha=0.7)
plt.title('Estimated PMF of X (Number of times heads > tails in 100 tosses)')
plt.xlabel('X (Number of times heads > tails)')
plt.ylabel('Probability')
plt.show()
```

```python
import numpy as np

num_simulations = 10000  # Number of simulations
num_tosses = 100  # Number of coin tosses

X_values = []
Y_values = []

for _ in range(num_simulations):
    heads = 0
    tails = 0
    X = 0  
    Y = 0 
    
    for _ in range(num_tosses):
        toss = np.random.choice(['H', 'T']) 
        
        if toss == 'H':
            heads += 1
        else:
            tails += 1
        
        if heads == tails:
            X += 1
        
        if heads > tails:
            Y += 1
    
    X_values.append(X)
    Y_values.append(Y)

X_values = np.array(X_values)
Y_values = np.array(Y_values)

E_X = np.mean(X_values)
E_Y = np.mean(Y_values)

print(f"Estimated Expected Value of X (times heads equals tails): {E_X:.4f}")
print(f"Estimated Expected Value of Y (times heads > tails): {E_Y:.4f}")
```
